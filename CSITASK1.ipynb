```python
df_cleaned = df.drop_duplicates(subset=['FirstName', 'LastName', 'HireDate'])
```

```python
#One-Hot Encoding: Converts categories into binary columns (0 or 1).
df_cleaned = pd.get_dummies(df, columns=['Department', 'JobRole'], drop_first=True)
```

```python
#Combine two or more features to capture interaction .
df['YearsWithManager_PerSalary'] = df['YearsWithCurrManager'] / df['Salary']
```

```python
# Handle missing values (example: filling missing values with mean or mode)
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Convert categorical variables into numerical ones (example: Gender)
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Convert the 'HireDate' column to datetime format
# Convert the 'HireDate' column to datetime format, letting pandas infer the format
df['HireDate'] = pd.to_datetime(df['HireDate'], dayfirst=True)
```

```python
# Calculate average salary by Department
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_by_dept)

# Calculate average salary by Gender
avg_salary_by_gender = df.groupby('Gender')['Salary'].mean()
print(avg_salary_by_gender)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# Histograms for key numerical variables
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

df['Salary'].hist(bins=20)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

```

```python
df['YearsAtCompany'].hist(bins=20)
plt.title('Years at Company Distribution')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')
plt.show()

```

```python
# Box plots for key numerical variables
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='JobRole', y='YearsAtCompany', data=df)
plt.title('Years at Company by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Years at Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Summary statistics in a report
report = {
    'Total Employees': len(df),
    'Average Age': df['Age'].mean(),
    'Average Salary': df['Salary'].mean(),
    'Attrition Rate': df['Attrition'].value_counts(normalize=True)['Yes'] * 100
}

print("Report Summary:")
for key, value in report.items():
    print(f"{key}: {value}")
```
